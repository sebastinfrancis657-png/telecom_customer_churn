# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
```

```python
num_cols=["Age","Tenure","MonthlyCharges","TotalCharges"]

df[num_cols]=scaler.fit_transform(df[num_cols])
df[num_cols]
```

```python
df
```

# Train-Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop("Churn", axis=1)
x
```

```python
y=df["Churn"]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=42)
```

# Train your first ML model (Logistic Regression)

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(max_iter=1000)

model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
y_pred = model.predict(x_test)
```

```python
print("Accuracy :", accuracy_score(y_test,y_pred))
print("Confusion matrix:", confusion_matrix(y_test,y_pred))
print("Clasiificaton report:", classification_report(y_test,y_pred))
```

# Solution 1: Use class_weight='balanced'

```python
model=LogisticRegression(max_iter=1000, class_weight="balanced")

model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print("Accuracy :", accuracy_score(y_test,y_pred))
print("Confusion matrix:", confusion_matrix(y_test,y_pred))
print("Clasiificaton report:", classification_report(y_test,y_pred))
```

# Random Forest (Stronger Model)

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf= RandomForestClassifier(n_estimators=100, random_state=42)

rf.fit(x_train,y_train)

y_pred_rf= rf.predict(x_test)
```

```python
print("Accuracy :", accuracy_score(y_test,y_pred_rf))
print("Confusion matrix:", confusion_matrix(y_test,y_pred_rf))
print("Clasiificaton report:", classification_report(y_test,y_pred_rf))
```

# Improve Random Forest

```python
rf= RandomForestClassifier(n_estimators=100, random_state=42, class_weight="balanced")

rf.fit(x_train,y_train)

y_pred_rf= rf.predict(x_test)
```

```python
print("Accuracy :", accuracy_score(y_test,y_pred_rf))
print("Confusion matrix:", confusion_matrix(y_test,y_pred_rf))
print("Clasiificaton report:", classification_report(y_test,y_pred_rf))
```

## Technique: Adjust Decision Threshold

```python
y_pred_rf = model.predict(x_test)
```

```python
y_prob_rf = rf.predict_proba(x_test)[:, 1]    #Get Probabilities
```

```python
y_pred_custom = (y_prob_rf > 0.3).astype(int)    #Apply Custom Threshold
```

```python
print("Accuracy :", accuracy_score(y_test,y_pred_custom))
print("Confusion matrix:", confusion_matrix(y_test,y_pred_custom))
print("Clasiificaton report:", classification_report(y_test,y_pred_custom))
```
